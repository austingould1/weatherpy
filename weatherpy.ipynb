```python
import pandas 
import matplotlib.pyplot as plt
import requests as req
import json
import random 
import numpy
from citipy import citipy
import time
```

```python
lat = numpy.random.uniform(low=-90.000, high=90.000, size=1200)
long = numpy.random.uniform(low=-180.000, high=180.000, size=1200)
lat_long = []
lat_long = zip(lat, long)
```

```python
cities = []
for coordinate_pair in lat_long:
    city = citipy.nearest_city(coordinate_pair[0], coordinate_pair[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
key= "f8149979e8b04cbe94b95b13ee1b2134"
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + key 
```

```python
# Your objective is to build a series of scatter plots to showcase the following relationships:

# * Temperature (F) vs. Latitude
# * Humidity (%) vs. Latitude
# * Cloudiness (%) vs. Latitude
# * Wind Speed (mph) vs. Latitude

count=0


temperature=[]
humidity=[]
clouds=[]
wind=[]
latt=[]
for city in cities:
    city_url = url +"&q=" + city
    print(city_url)
    response = req.get(city_url).json()
    count +=1
    try:
        temp = response["main"]["temp"]
        temperature.append(temp)
        humid = response["main"]["humidity"]
        humidity.append(humid)
        Cloudiness = response["clouds"]["all"]
        clouds.append(Cloudiness)
        wind_speed = response["wind"]["speed"]
        wind.append(wind_speed)
        lat = response["coord"]["lat"]
        latt.append(lat)
        

    except:
        print("City data not found")
    if count==60:
        count=0
        time.sleep(60)
        
    
```

```python
plt.scatter(x = latt, y =temperature, alpha=.5)
    
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.savefig('lat_temp.png')
plt.show()
```

```python
plt.scatter(x = latt, y =humidity, alpha=.5)
    
plt.xlabel("latitude")
plt.ylabel("humidity")
plt.savefig('lat_hum.png')
plt.show()
```

```python
plt.scatter(x = latt, y =clouds, alpha=.5)
    
plt.xlabel("latitude")
plt.ylabel("clouds")
plt.savefig('lat_cloud.png')
plt.show()
```

```python
plt.scatter(x = latt, y =wind, alpha=.5)
    
plt.xlabel("latitude")
plt.ylabel("wind")
plt.savefig('lat_wind.png')
plt.show()
```

```python
# * Save both a CSV of all data retrieved and png images for each scatter plot.
district_summary_df = pandas.DataFrame({ 
    "temperature": temperature,
    "humidity": humidity,
    "clouds": clouds,
    "wind": wind, 
    "latitude": latt
    
})
district_summary_df.head()

```
